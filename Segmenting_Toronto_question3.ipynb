```python
from bs4 import BeautifulSoup
import pandas as pd
```

Read the html file using Beautifulsoup. Look for content within table, search for all the rows in the table that are separated by <tr>. Assign the first row as the column names of the data frame, and assign the other rows to the rest of the dataframe. 

```python
with open("ListofpostalcodesofCanada.html") as fp:
    soup = BeautifulSoup(fp, "lxml")
df= pd.DataFrame()
i=0
for table in soup.table.find_all('tr'):
    row=table.text.split('\n')[1:4]
    if i==0:
        df=pd.DataFrame(columns=row)
    else:
        df.loc[i] = row
    i=i+1

```

Remove all the rows in the data frame that has value "Not assigned" in the "Borough" column.

```python
df_nh=df[df.Borough!='Not assigned'].sort_values(by=['Postcode']).reset_index(drop=True)
```

Assign the name of Borough to the name of Neighborhood that has the value "Not assigned".

```python
index=df_nh[df_nh['Neighbourhood']=='Not assigned'].index.values.tolist()
df_nh.iloc[index,2]=df_nh.iloc[index,1]
```

Combine all the rows with the same Postcode into the same row, with Neighborhood name listed togeter and separated by ",".

```python
nhlist=df_nh.groupby('Postcode')['Neighbourhood'].apply(list)
df_nh.drop_duplicates(subset='Postcode', keep='first',inplace=True)
j=0
for nh in nhlist:
    nhname = ','.join(nh)
    df_nh.iloc[j,2]=nhname  
    j=j+1

df_nh.reset_index(drop=True)
```

Open the CSV file that contains the latitude and longitude of each postcode area, and change the first column name to match with the dataframe above.

```python
df_code =pd.read_csv('Geospatial_Coordinates.csv')
df_code.columns=['Postcode', 'Latitude', 'Longitude']
```

Join the two dataframe using the postcode as key.

```python
df_join=df_nh.join(df_code.set_index('Postcode'), on='Postcode')
df_join.reset_index(drop=True)
```

Borough names that contains the word Toronto.

```python
df_tr=df_join[df_join.Borough.str.contains('Toronto')]
df_tr.reset_index(drop=True)
```

Create function to find venues in the neighborhood and venue categroy. 

```python
def getnhvenues(name, latitude, longitude, radius=500):
    
    venue_lists=[]
    for na, lat, lng in zip(name, latitude, longitude):
        print(na)
        
        url='https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}'.format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        lng,
        radius)
        
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venue_lists.append([(
        na,
        lat,
        lng,
        v['venue']['name'],
        v['venue']['location']['lat'],
        v['venue']['location']['lng'],
        v['venue']['categories'][0]['name']) for v in results])
        
    nhvenues=pd.DataFrame([item for venue_lists in venue_lists for item in venue_lists])
    nhvenues.columns=['Neighbourhood',
                     'Latitude',
                     'Longitude',
                     'Venue',
                     'Venue Lat',
                     'Venue Lng',
                     'Venue Cat']
    
    return(nhvenues)
```

Get venues in all borough.

```python
tr_venues = getnhvenues(name=df_tr['Neighbourhood'],
                       latitude=df_tr['Latitude'],
                       longitude=df_tr['Longitude'])
```

```python
print(tr_venues.shape)
tr_venues.head()
tr_venues.groupby('Neighbourhood').count()
print('There are {} unique categories.'.format(len(tr_venues['Venue Cat'].unique())))
```

One hot coding all the venue categories, and calculate the frequencies of each category in all neighborhoods.

```python
tr_onehot=pd.get_dummies(tr_venues[['Venue Cat']], prefix="", prefix_sep="")
tr_onehot['Neighbourhood']=tr_venues['Neighbourhood']

fixed_columns =[tr_onehot.columns[-1]]+list(tr_onehot.columns[:-1])
tr_onehot=tr_onehot[fixed_columns]

tr_grouped=tr_onehot.groupby('Neighbourhood').mean().reset_index()
tr_grouped.head()
```

Function to sort the venues in descending order

```python
def return_most_common_venues(row, num_top_venues):
    row_cat = row.iloc[1:]
    row_cat_sorted = row_cat.sort_values(ascending=False)
    return row_cat_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues=10

indicators=['st', 'nd', 'rd']

columns=['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} most common venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{} most common venue'.format(ind+1))
        

nh_venues_sorted = pd.DataFrame(columns=columns)
nh_venues_sorted['Neighbourhood']=tr_grouped['Neighbourhood']

print(tr_grouped.shape[0])
for ind in np.arange(tr_grouped.shape[0]):
    nh_venues_sorted.iloc[ind,1:]=return_most_common_venues(tr_grouped.iloc[ind,:], num_top_venues)
    
nh_venues_sorted.head()
```

Run k-means to cluster the neighborhoods into 5 clusters.

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

kcluster=6

tr_grouped_cluster=tr_grouped.drop('Neighbourhood',1)

kmeans = KMeans(n_clusters=kcluster, random_state=0).fit(tr_grouped_cluster)

print(kmeans.labels_)

#nh_venues_sorted.insert(0, 'Cluster labels', kmeans.labels_)

tr_merged = df_tr
tr_merged=tr_merged.join(nh_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')
tr_merged.sort_values(by=['Cluster labels'])
```

Visualize the different clusters on the map.

```python
map_clusters = folium.Map(location=[lattr,lngtr], zoom_start=11)

x = np.arange(kcluster)
y = [i+x+(i*x)**2 for i in range(kcluster)]
color_array = cm.rainbow(np.linspace(0,1,kcluster))
rainbow=[colors.rgb2hex(i) for i in color_array]

markers_colors=[]

for lat, lng, nh, cluster in zip(tr_merged['Latitude'], tr_merged['Longitude'], tr_merged['Neighbourhood'], tr_merged['Cluster labels']):
    label=folium.Popup(str(nh)+' Cluster '+str(cluster), parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color=rainbow[cluster-1],
    fill=True,
    fill_color=rainbow[cluster-1],
    fill_opacity=0.6).add_to(map_clusters)
    
map_clusters
```

The neighborhoods in Toronto containing the name "Toronto" can be mainly divided into 3 clusters. Cluster 0 (8 postcodes red in the map) contains neighborhoods that have the most coffee shops. Cluster 1 (3 postcodes purple in the map) contains neighborhoods that have the most park or gym. Cluster 2 (24 postcodes  blue in the map) contains neighborhoods that have the most restaurants. Cluster 3, 4, and 5 (1 postcode each, cyan, lime, and orange in map) are neighborhoods that do not fit well into the first three clusters.
